```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load data
df = pd.read_csv("patients.csv")

# Step 3: Inspect data (optional)
df.head()
```

```python
# Calculate average weight loss per program
avg_loss = df.groupby('program')['weight_loss'].mean()

# Create bar chart
plt.figure(figsize=(6,4))
sns.barplot(x=avg_loss.index, y=avg_loss.values, palette="Blues_d")
plt.title("Average Weight Loss per Program", fontsize=14)
plt.xlabel("Program")
plt.ylabel("Average Weight Loss (lbs)", fontsize=12)
plt.tight_layout()
plt.show()
```
