```python
import pandas as pd

# Load the CSV
df = pd.read_csv('patients.csv')

# See the first few rows
df.head()
```

```python
df.columns
```

```python
avg_loss = df.groupby('program')['weight_loss'].mean()
```

```python
avg_loss = df.groupby('program')['weight_loss'].mean()
avg_loss
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.barplot(x=avg_loss.index, y=avg_loss.values)
plt.title("Average Weight Loss per Program")
plt.xlabel("Program")
plt.ylabel("Average Weight Loss")
plt.show()
```
